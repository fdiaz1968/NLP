### R — Similitud sobre DFM TF-IDF

```r
library(quanteda)
library(quanteda.textstats)

textos <- c(
  doc1 = "El Banco Central subió la tasa de interés ante el aumento de la inflación.",
  doc2 = "Los inversionistas reaccionaron con cautela frente a la volatilidad del mercado.",
  doc3 = "La economía mostró señales de recuperación durante el último trimestre."
)

dfm_tfidf_nlp <- corpus(textos) |>
  tokens(remove_punct = TRUE) |>
  tokens_remove(stopwords("spanish")) |>
  dfm() |>
  dfm_tfidf()

textstat_simil(dfm_tfidf_nlp, method = "cosine")
```

### Python — Embeddings semánticos, clustering y MDS

````python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.manifold import MDS
from scipy.cluster.hierarchy import linkage, dendrogram

modelo = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")

oraciones = [
    "El Banco Central subió la tasa de interés ante el aumento de la inflación.",
    "Los inversionistas reaccionaron con cautela frente a la volatilidad del mercado.",
    "La economía mostró señales de recuperación durante el último trimestre.",
]

embeddings = modelo.encode(oraciones)
similitud = cosine_similarity(embeddings)
print(similitud)

# Clustering jerárquico
enlaces = linkage(embeddings, method="average", metric="cosine")
dendrogram(enlaces, labels=[f"doc{i+1}" for i in range(len(oraciones))])

# Escalamiento multidimensional (MDS) sobre distancias 1 - similitud coseno
mds = MDS(n_components=2, dissimilarity="precomputed", random_state=42)
coordenadas = mds.fit_transform(1 - similitud)
print(coordenadas)
````
