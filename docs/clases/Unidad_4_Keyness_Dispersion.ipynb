## Desarrollo

### R — Keyness estadístico

```r
library(quanteda)
library(quanteda.textstats)
library(quanteda.textplots)

textos <- c(
  doc1 = "El Banco Central subió la tasa de interés ante el aumento de la inflación.",
  doc2 = "Los inversionistas reaccionaron con cautela frente a la volatilidad del mercado.",
  doc3 = "La economía mostró señales de recuperación durante el último trimestre."
)

dfm_grupos <- corpus(textos, docvars = data.frame(fuente = c("prensa", "prensa", "informe"))) |>
  tokens(remove_punct = TRUE) |>
  tokens_remove(stopwords("spanish")) |>
  dfm() |>
  dfm_group(groups = fuente)

# Keyness de "prensa" respecto al resto del corpus
keyness <- textstat_keyness(dfm_grupos, target = "prensa")
head(keyness)

textplot_keyness(keyness)
textplot_xray(kwic(tokens(corpus(textos)), pattern = "inflación"))
```

### Python — Narrativa automática con un LLM

````python
from transformers import pipeline

generador = pipeline("text-generation", model="datificate/gpt2-small-spanish")

terminos_clave = ["inflación", "tasa", "inversionistas"]
prompt = (
    "Explica en un párrafo por qué estos términos son clave en el corpus: "
    + ", ".join(terminos_clave)
)

narrativa = generador(prompt, max_new_tokens=80, do_sample=True)[0]["generated_text"]
print(narrativa)
````
