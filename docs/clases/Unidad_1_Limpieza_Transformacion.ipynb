## 🌐 ¿Qué es el Procesamiento de Lenguaje Natural (NLP)?

El **Procesamiento de Lenguaje Natural (Natural Language Processing, NLP)** es un área interdisciplinaria que combina lingüística, estadística, aprendizaje automático e inteligencia artificial para permitir que las computadoras analicen, interpreten y generen lenguaje humano.

Tradicionalmente, el NLP se basaba en representaciones estadísticas del texto, tales como frecuencias de palabras, matrices documento-término y modelos probabilísticos. En los últimos años, el desarrollo de los **Large Language Models (LLM)** y de las arquitecturas **Transformer** ha ampliado significativamente las capacidades del campo, permitiendo comprender relaciones semánticas complejas, generar texto y realizar tareas de razonamiento sobre lenguaje.

El objetivo central del NLP sigue siendo el mismo: **transformar información textual no estructurada en conocimiento útil y analizable**. Sin embargo, hoy disponemos de múltiples enfoques para lograrlo, cada uno con ventajas y limitaciones.

---

## 🛠️ Aplicaciones del NLP que veremos en este curso

Durante este curso estudiaremos tanto enfoques clásicos como modernos de NLP utilizando **R**, **Python** y modelos basados en **Transformers**.

- Similitud semántica mediante *sentence-transformers*.
- Descubrimiento de tópicos con BERTopic.
- Uso de LLM para interpretación, codificación y análisis asistido de textos.

## 🤖 NLP Tradicional vs Modelos de Lenguaje de Gran Escala (LLMs)

### 🧠 ¿Qué es el NLP?

El **Procesamiento de Lenguaje Natural (NLP)** es un campo de la inteligencia artificial que permite a las computadoras analizar, comprender y generar lenguaje humano. Combina herramientas estadísticas, lingüísticas y computacionales para transformar texto no estructurado en información útil y analizable.

Entre sus aplicaciones más comunes se encuentran:

- Limpieza y transformación de texto.
- Análisis de frecuencia y coocurrencia.
- Construcción de matrices documento-término.
- Modelamiento de tópicos (LDA).
- Análisis de sentimiento.
- Medición de similitud entre documentos.
- Clasificación automática de textos.

---

### 🚀 ¿Qué son los LLMs?

Los **Modelos de Lenguaje de Gran Escala (Large Language Models, LLMs)** son modelos de inteligencia artificial basados en la arquitectura **Transformer**, entrenados sobre enormes colecciones de texto.

Gracias a este entrenamiento, pueden capturar relaciones semánticas complejas y realizar múltiples tareas relacionadas con el lenguaje, tales como:

- Responder preguntas.
- Generar texto.
- Resumir documentos.
- Traducir entre idiomas.
- Clasificar textos.
- Extraer información relevante.

Modelos como GPT, Claude, Gemini, Llama o Mistral representan algunos de los ejemplos más conocidos de esta nueva generación de herramientas.

---

### ✅ ¿Para qué es mejor cada enfoque?

#### NLP Tradicional

Es especialmente útil cuando:

- Se requiere transparencia e interpretabilidad.
- Se desea controlar cada etapa del procesamiento.
- Se trabaja con corpus grandes y bien estructurados.
- Se busca máxima reproducibilidad metodológica.
- Se dispone de recursos computacionales limitados.

#### LLMs

Son especialmente útiles cuando:

- Se requiere comprensión contextual profunda.
- Los textos contienen ambigüedad, jerga o lenguaje complejo.
- Se necesitan tareas de generación de texto o resumen automático.
- No se dispone de datos etiquetados para entrenamiento.
- Se busca flexibilidad para resolver múltiples tareas con un mismo modelo.

---

### 📊 Comparación de Ventajas y Desventajas

| Característica | NLP Tradicional | LLMs |
|----------------|----------------|-------|
| Interpretabilidad | Alta | Baja |
| Requerimientos computacionales | Bajos | Altos |
| Reproducibilidad | Alta | Moderada |
| Escalabilidad | Muy alta | Alta, pero costosa |
| Comprensión contextual | Limitada | Muy alta |
| Necesidad de datos etiquetados | Frecuentemente necesaria | Puede operar con *zero-shot* o *few-shot* |
| Control metodológico | Alto | Menor |
| Costo de implementación | Bajo | Variable |

---

### 🎯 Conclusión

Los enfoques tradicionales y los LLMs deben entenderse como herramientas complementarias más que como alternativas excluyentes.

Los métodos clásicos continúan siendo fundamentales para tareas descriptivas, análisis reproducibles y estudios académicos rigurosos. Los LLMs, por su parte, ofrecen capacidades superiores para comprender contexto, interpretar significado y automatizar tareas complejas.

# I. Cargando **"quanteda"** y nuestro primer ***corpus***

## 📦 ¿Qué es `quanteda`?

`quanteda` (Quantitative Analysis of Textual Data) es un **paquete de R** diseñado para facilitar el análisis cuantitativo de textos. Es una herramienta poderosa, eficiente y flexible que permite transformar texto no estructurado en objetos estructurados para su análisis estadístico.

---

### 🧰 Funcionalidades principales

- **Creación y manipulación de corpus**  
  Permite importar, visualizar y gestionar grandes colecciones de textos.

- **Tokenización y limpieza de texto**  
  Remueve puntuación, stopwords, números, símbolos y permite hacer stemming y lematización.

- **Construcción de la Document-Feature Matrix (DFM)**  
  Convierte el texto en una matriz estructurada donde las filas son documentos y las columnas son términos.

- **Análisis de frecuencia y coocurrencia**  
  Mide cuán frecuentemente ocurren palabras o aparecen juntas.

- **Visualizaciones**  
  Incluye funciones para generar nubes de palabras, gráficos de dispersión léxica, keyness, etc.

- **Modelado de tópicos y análisis avanzado**  
  Se puede integrar fácilmente con otros paquetes como `topicmodels`, `tidytext`, o `stm`.

---

### 💡 ¿Por qué usar `quanteda`?

- Optimizado para **grandes volúmenes de texto**
- Basado en estructuras de datos eficientes (`dfm`)
- Compatible con el **ecosistema tidyverse**
- Ideal para investigadores en **ciencias sociales, lingüística computacional, economía y comunicación**

---

### 🧪 Ejemplos de aplicaciones (en este curso)

- Análisis de sentimiento en reseñas de películas
- Identificación de tópicos con LDA
- Detección de palabras clave por grupo (keyness)
- Visualización de patrones léxicos y semánticos
- Comparación de documentos por similitud

> En resumen, `quanteda` es una de las herramientas más completas en R para realizar NLP clásico de forma transparente, reproducible y eficiente.

## 📚 ¿Qué es un *Corpus*?

En lingüística y procesamiento del lenguaje natural (NLP), un **corpus** es una colección organizada de documentos textuales que constituye la unidad básica de análisis. Los documentos pueden provenir de múltiples fuentes, tales como noticias, artículos académicos, redes sociales, informes financieros, documentos legales o transcripciones de entrevistas.

Los corpus permiten estudiar patrones lingüísticos, extraer información y entrenar modelos capaces de analizar o generar lenguaje humano.

---

### 🧩 Características de un corpus

- Está compuesto por uno o más documentos.
- Puede contener textos escritos o transcripciones de lenguaje hablado.
- Permite analizar el lenguaje en contextos reales.
- Puede estar acompañado de **metadatos** (*document variables* o *docvars*), tales como:
  - Fecha de publicación.
  - Autor o fuente.
  - Categoría temática.
  - País o ubicación.
  - Etiquetas de sentimiento o clasificación.

---

### 🛠️ ¿Para qué sirve un corpus?

Los corpus constituyen la materia prima de prácticamente todas las aplicaciones modernas de NLP:

- Análisis de frecuencias y vocabulario.
- Identificación de patrones lingüísticos.
- Análisis de sentimiento.
- Modelamiento de tópicos.
- Medición de similitud entre documentos.
- Entrenamiento y evaluación de modelos de clasificación.
- Construcción de embeddings y aplicaciones basadas en LLMs.

Sin un corpus adecuadamente construido y documentado, incluso los modelos más sofisticados producirán resultados poco confiables.

Los **metadatos** son *datos sobre los datos lingüísticos*. Es decir, no forman parte del contenido textual principal, sino que proporcionan **información adicional que describe, estructura o contextualiza ese texto**.

# **II. Tokenization**

## ✂️ ¿Qué es la Tokenización?

La **tokenización** es un proceso fundamental en el procesamiento del lenguaje natural (NLP) que consiste en dividir un texto en unidades más pequeñas llamadas **tokens**.

Un **token** puede ser:

- Una palabra.
- Un número.
- Un signo de puntuación.
- Un fragmento de palabra.
- O cualquier otra unidad significativa del texto.

---

### 🎯 Objetivo

Transformar el texto crudo en una representación estructurada que pueda ser procesada por algoritmos computacionales.

---

### 🧠 ¿Por qué es importante?

- Es el primer paso de prácticamente todos los flujos de trabajo de NLP.
- Permite limpiar y normalizar textos.
- Facilita el cálculo de frecuencias, similitudes y representaciones vectoriales.
- Constituye la base para construir matrices documento-término (DFM), embeddings y modelos de lenguaje.

---

### 🤖 ¿Los LLM utilizan tokenización?

Sí. Los modelos modernos basados en Transformers también tokenizan el texto antes de procesarlo.

La diferencia es que los tokenizadores tradicionales suelen trabajar con palabras completas, mientras que los LLM utilizan frecuentemente **"subpalabras"** (*subword tokens*), lo que les permite manejar vocabulario muy amplio, palabras desconocidas y múltiples idiomas.

Por ejemplo:

```text
Texto:
"La empresa aumentó sus utilidades"

Tokenización tradicional:
["la", "empresa", "aumentó", "sus", "utilidades"]

Tokenización de un LLM:
["La", "empresa", "aument", "ó", "sus", "util", "idades"]
```

---

📌 En **Quanteda**, la tokenización se realiza habitualmente mediante:

```r
tokens(corpus)
```

y constituye el paso previo a la construcción de una matriz documento-término (DFM).

---

## Un Primer Ejemplo






Consideremos el siguiente texto:

"Mario es un gran empresario. Creó su propia empresa a partir de un modesto emprendimiento."

````r
my_text <- c(
"Mario es un gran empresario. Creó su propia empresa a partir de un modesto emprendimiento."
)
````

Generamos un *Corpus*

````r
my_corpus <- corpus(my_text)
print(my_corpus)
class(my_corpus)
````

Aplicamos la función *toks()* al corpus:

````r
toks <- tokens(my_corpus)
print(toks)
class(toks)
````

Podemos cambiar todas las palabras a minúsculas - *tokens_tolower()*:

````r
toks_lower <- tokens_tolower(toks)
print(toks_lower)
````

Y elminar *stop words* y puntuación:

````r
toks_nopunct_stop <-tokens(toks_lower, remove_punct = TRUE) |>
  tokens_remove(pattern = stopwords("es"))

toks_nopunct_stop 
````

Podemos hacer *stemming* (el stemming es una técnica para reducir una palabra a su raíz o “stem”, es decir, a una forma básica que representa todas sus variantes *flexionadas*.)

````r
toks_nopunct_stop_stem <- tokens_wordstem(toks_nopunct_stop)
toks_nopunct_stop_stem
````

### ⚠️ El stemming por defecto no funciona en español

El resultado anterior no se ve como una raíz razonable: `tokens_wordstem()` usa por defecto el algoritmo de Porter para **inglés**, por lo que palabras como *"empresario"* o *"emprendimiento"* no se reducen correctamente.

Para aplicar stemming en español (y en muchos otros idiomas), `quanteda` se apoya en el paquete **SnowballC**, que implementa el algoritmo de Snowball para múltiples idiomas.


````r
#install.packages("SnowballC")
library(SnowballC)
````

````r
toks_nopunct_stop_stem <- tokens_wordstem(toks_nopunct_stop, language = "spanish")
toks_nopunct_stop_stem
````

---

## 🌿 Stemming vs. Lematización

- **Stemming**: trunca la palabra a una raíz aproximada, que muchas veces no es una palabra real del idioma.
  Ejemplo (en nuestra frase): `empresario` → `empresari`, `empresa` → `empres`, `emprendimiento` → `emprend`.

- **Lematización**: reduce cada palabra a su **forma de diccionario** (lema): el infinitivo de los verbos, el masculino singular de los adjetivos, etc.
  Ejemplo (en nuestra frase): `creó` → `crear` (conjugación → infinitivo), `gran` → `grande` (apócope → forma plena), `propia` → `propio` (femenino → forma de diccionario).

`quanteda` no incluye un lematizador propio, porque lematizar correctamente requiere conocer la **categoría gramatical (POS)** de cada palabra. Para eso existen paquetes especializados:

- **R**: `udpipe`, `spacyr`
- **Python**: `spaCy`, `stanza`

Para ilustrar la diferencia conceptual sin instalar modelos pesados, construimos a mano un pequeño **diccionario de lematización** y lo aplicamos con `tokens_replace()`:


````r
# Diccionario de lematización manual (solo para ilustrar el concepto)
lemma_dict <- c(
  "creó" = "crear",
  "gran" = "grande",
  "propia" = "propio"
)

toks_lemma <- tokens_replace(
  toks_nopunct_stop,
  pattern = names(lemma_dict),
  replacement = lemma_dict,
  valuetype = "fixed"
)

print(toks_nopunct_stop_stem)  # stemming
print(toks_lemma)              # lematización (manual)
````

> 🔍 **Comparación**: el stemming reduce `empresario`, `empresa` y `emprendimiento` a raíces como `empresari`, `empres` y `emprend`, que no son palabras del español. La lematización, en cambio, normaliza `creó` → `crear`, `gran` → `grande` y `propia` → `propio`: formas que sí existen en el diccionario, pero que requieren reglas (o un modelo lingüístico) capaces de reconocer conjugaciones verbales, apócopes y concordancia de género/número.
>
> En un corpus real, este diccionario se construiría automáticamente con `udpipe::udpipe_annotate()`, que etiqueta cada token con su lema y su categoría gramatical (POS), en lugar de definirse a mano.


---

### 🤖 ¿Y qué ocurre en los LLM?

Los modelos modernos basados en **Transformers** normalmente no requieren aplicar *stemming* o *lematización* antes del análisis.

Esto se debe a que utilizan mecanismos de tokenización basados en **subpalabras** y representaciones contextuales (*embeddings*), capaces de capturar similitudes entre términos relacionados sin necesidad de reducir explícitamente cada palabra a una raíz común.

Por ejemplo, palabras como:

```text
empresa
empresas
empresarial
empresario
```

suelen ser interpretadas por un LLM como conceptos estrechamente relacionados, incluso sin aplicar stemming.

> 📌 El stemming sigue siendo muy útil en enfoques tradicionales basados en frecuencias, matrices documento-término (DFM) y TF-IDF, donde reducir la variabilidad léxica puede mejorar la calidad del análisis.

## ✂️ Tokens, Guiones y Palabras Compuestas

Cuando analizamos texto con `quanteda`, es común encontrar **palabras compuestas con guiones**, como:

> `"Fellow-Citizens"`, `"God-given"`, `"non-believers"`, `"self-reliance"`, `"self-supporting"`, `"well-spent"`, `"well-regulated"`...

---

### 🔧 `tokens()` y el argumento `split_hyphens`

El argumento `split_hyphens` de la función `tokens()` controla si las palabras con guiones deben tratarse como un solo token o dividirse:

- `split_hyphens = FALSE` (por defecto): mantiene las palabras con guiones como **un solo token**.  
  Ejemplo: `"God-given"` → `"God-given"`

- `split_hyphens = TRUE`: divide las palabras compuestas en múltiples **tokens separados**, eliminando el guion.  
  Ejemplo: `"God-given"` → `"God"` + `"given"`

---

### 🎯 ¿Por qué importa?

- Afecta el **conteo de frecuencia** de términos.
- Impacta el resultado de técnicas como **stemming**, **keyness**, o **modelamiento de tópicos**.
- Puede ser útil para capturar **patrones léxicos** de prefijos como `non-`, `self-`, `well-`, etc.

> 🔍 La decisión de dividir o mantener palabras compuestas debe basarse en los **objetivos del análisis**.


---

### 🤖 ¿Qué ocurre en los LLM?

Los modelos modernos basados en **Transformers** también tokenizan las palabras compuestas, pero normalmente no requieren que el investigador decida explícitamente si un término debe dividirse o mantenerse unido.

Por ejemplo:

```text
self-reliance
```

puede ser segmentado internamente en varias subpalabras o símbolos según el tokenizador utilizado por el modelo.

Gracias a los **embeddings contextuales**, los LLM suelen conservar la relación semántica entre los componentes de la palabra compuesta sin necesidad de reglas manuales como `split_hyphens = TRUE`.

> 📌 En NLP tradicional, la decisión de dividir o mantener palabras compuestas puede afectar significativamente las frecuencias, la DFM y los modelos posteriores. En los LLM, esta decisión suele ser manejada automáticamente por el tokenizador del modelo.

---

## 🔠 Mayúsculas, Signos de Puntuación y *Stop Words*

Una vez que hemos tokenizado los textos, es común aplicar transformaciones adicionales para normalizar los datos y reducir la variabilidad léxica.

### 🔡 Conversión a minúsculas

Usamos `tokens_tolower()` para estandarizar las palabras y evitar que términos como `"Government"` y `"government"` se traten como elementos diferentes.

### ❌ Eliminación de puntuación y símbolos

Al tokenizar, podemos utilizar los argumentos `remove_punct = TRUE` y `remove_symbols = TRUE` para eliminar caracteres como comas, puntos, signos y símbolos especiales.

### 🧹 Eliminación de *Stop Words*

Las *stopwords* (como `"the"`, `"and"` o `"in"`) son palabras muy frecuentes que suelen aportar poca información temática. Mediante `tokens_remove()` pueden eliminarse para reducir ruido y simplificar el análisis.

---

### 🎯 ¿Por qué realizar estas transformaciones?

En los enfoques tradicionales basados en frecuencias, matrices documento-término (DFM) y TF-IDF, estas operaciones permiten:

- Reducir la dimensionalidad de los datos.
- Concentrar la información en términos más relevantes.
- Mejorar la interpretación de frecuencias y tópicos.
- Facilitar la comparación entre documentos.

Por ello, constituyen una práctica habitual en muchos flujos de trabajo de NLP clásico.

---



> 📌 En NLP tradicional estas transformaciones suelen mejorar los resultados al reducir la dimensionalidad de los datos. En los LLM, una limpieza excesiva puede eliminar información valiosa para la comprensión del contexto.
---


---

### 🤖 ¿Qué ocurre en los LLM?

En los enfoques tradicionales basados en frecuencias y matrices documento-término (DFM), estas transformaciones suelen mejorar el análisis al reducir la variabilidad léxica y disminuir la dimensionalidad de los datos.

Sin embargo, los modelos modernos basados en **Transformers** normalmente trabajan con el texto casi en su forma original. Las mayúsculas, la puntuación y muchas palabras funcionales pueden contener información relevante para interpretar correctamente el contexto.

Eliminar estos elementos puede:

- **Distorsionar el significado** (por ejemplo, eliminar “no” en “no feliz” cambia completamente el sentido),
- **Romper la sintaxis**, lo que dificulta la comprensión del contexto por parte del modelo,
- **Reducir el rendimiento** en tareas como clasificación, análisis de sentimiento, QA, etc.

Por esta razón, los LLM suelen conservar gran parte de la información original del texto y requieren mucho menos preprocesamiento que los enfoques clásicos.

> 📌 Una regla práctica es que mientras los métodos tradicionales suelen beneficiarse de una limpieza agresiva del texto, los Transformers y LLM generalmente funcionan mejor con textos cercanos a su forma original.

### ¿Por qué eliminar las "stop words"?



1.   Reducción del ruido: Las stopwords son palabras muy comunes en un idioma determinado, como "el", "de", "y", "a", etc. Estas palabras suelen tener poca información semántica y, en muchos casos, no aportan un valor significativo al análisis de texto. Al eliminar las stopwords, se reduce el ruido en los datos y se enfoca en las palabras más relevantes para el análisis.

2.   Mejora de la eficiencia computacional: Las stopwords son palabras que aparecen con mucha frecuencia en los documentos. Al eliminarlas, se reduce la cantidad de palabras en el texto, lo que a su vez reduce la cantidad de cálculos y operaciones necesarias durante el análisis. Esto puede acelerar el procesamiento y mejorar la eficiencia computacional.


3.   Enfocarse en el contexto y las palabras clave: Al eliminar las stopwords, se puede poner mayor énfasis en las palabras que tienen más peso semántico y contribuyen significativamente al significado y al contexto de un texto. Esto puede ayudar a identificar y analizar las palabras clave, los patrones y las relaciones más relevantes en el análisis de texto.

Es importante destacar que la eliminación de stopwords no es aplicable en todos los casos y en todos los tipos de análisis de texto. En algunos casos, como el análisis de contexto o la identificación de temas generales, las stopwords pueden contener información relevante y no deben eliminarse. Por lo tanto, es esencial considerar el contexto y los objetivos del análisis antes de decidir si es apropiado eliminar o conservar las stopwords.

## 🌱 IV. Stemming

El **stemming** es el proceso de reducir las palabras a su forma básica o raíz léxica, llamada *stem*. Esta raíz no siempre es una palabra válida, pero permite agrupar variantes morfológicas como si fueran equivalentes. Por ejemplo:

- *run*, *runner*, *running* → **run**
- *connect*, *connected*, *connecting*, *connection* → **connect**

Este proceso es útil para:

- Reducir la dimensión del vocabulario.
- Identificar similitudes semánticas entre términos relacionados.
- Mejorar tareas como el análisis de frecuencia, modelamiento de tópicos o recuperación de información.

En `quanteda`, el stemming se aplica con la función `tokens_wordstem()`. Es especialmente útil cuando queremos enfocarnos en el contenido general de un texto, sin preocuparnos por sus variaciones gramaticales.

> ⚠️ A diferencia de la **lematización**, el stemming no devuelve necesariamente palabras del diccionario, sino formas truncadas útiles para análisis computacional.


---


````r
tokens_wordstem(toks_nopunct_stop_low)
````

En cuanto a las limitaciones inherentes al stemming,  probablemente el problema más importante está relacionado con el **over-stemming** (situación en la que dos palabras no relacionadas se reducen a la misma raíz) y el **under-stemming** (situación en la que dos palabras relacionadas se reducen a diferentes raíces). En este sentido, trabajar con la versión *stemmed* de un corpus requiere más intervención "humana" para que el analista pueda evaluar correctamente el significado correcto de una palabra **stemmed**.
(*over-stemming* y  *under-stemming* me recuerdan el trade-off entre *over-fitting* y *under-fitting* en modelos de ML).

# V. Key Words in Context - kwic()

### 📌 ¿Qué hace `kwic()`?

`kwic()` es una técnica clásica de análisis de texto que muestra una palabra clave rodeada por un número definido de palabras antes y después. Es útil para observar cómo se usa un término en distintos contextos lingüísticos.

````r
toks <- tokens(data_corpus_inaugural)
kw_inaug <- kwic(toks, pattern = c("liberty", "people"), window = 4)
print(tail(kw_inaug, 10))

````

Nótese que en este caso, se utiliza la versión más básica del texto tokenizado.  Al utilizar la versión básica del texto tokenizado, se pueden presentar las palabras clave junto con su contexto de manera más clara en un lenguage que hace más sencilla la interpretación humana.

### 🔍 ¿Pueden los LLMs realizar tareas tipo `kwic()`?

### ✅ Sí, pero no es su principal fortaleza.

Los **modelos de lenguaje grande (LLMs)** pueden identificar ocurrencias de palabras o expresiones dentro de un texto y mostrar el contexto que las rodea, realizando una tarea conceptualmente similar a `kwic()` (*Key Word In Context*).

Sin embargo, los LLMs fueron diseñados para comprender e interpretar lenguaje, no para realizar búsquedas estructuradas sobre corpus extensos.

---

### 🤖 ¿Qué puede hacer un LLM?

Un LLM puede:

- Identificar una palabra o expresión en un texto.
- Resumir el contexto en el que aparece.
- Explicar diferencias de significado según el contexto.
- Agrupar usos similares de un término.

Por ejemplo, un LLM puede responder preguntas como:

> "¿Cómo se utiliza la palabra *government* a lo largo de estos discursos?"

o

> "¿En qué contextos aparece la palabra *inflation* en este conjunto de documentos?"

---

### ⚠️ ¿Qué limitaciones tiene?

Para tareas de búsqueda sistemática sobre corpus grandes, los LLM presentan algunas limitaciones:

- Pueden omitir ocurrencias.
- No garantizan reproducibilidad perfecta.
- Son menos eficientes para recorrer miles de documentos.
- Resultan innecesariamente costosos para tareas puramente mecánicas de búsqueda.

---

### 📊 Comparación: LLM vs `kwic()`

| Aspecto | LLM | `kwic()` |
|----------|------|---------|
| Búsqueda exacta de ocurrencias | ⚠️ Buena, pero no garantizada | ✅ Exacta |
| Escalabilidad | ⚠️ Limitada por contexto y costo | ✅ Muy alta |
| Reproducibilidad | ⚠️ Moderada | ✅ Total |
| Interpretación semántica | ✅ Muy alta | ❌ Limitada |
| Velocidad en corpus grandes | ❌ Menor | ✅ Muy alta |
| Transparencia metodológica | ⚠️ Moderada | ✅ Alta |

---

### 🎯 ¿Cuándo usar cada uno?

Utiliza **`kwic()`** cuando necesites:

- Recuperar todas las ocurrencias de un término.
- Realizar análisis reproducibles.
- Trabajar con corpus grandes.
- Obtener resultados exactos y verificables.

Utiliza un **LLM** cuando necesites:

- Interpretar los contextos encontrados.
- Resumir patrones de uso.
- Comparar significados entre documentos.
- Explorar relaciones semánticas complejas.

> 📌 En la práctica, ambos enfoques son complementarios: `kwic()` permite recuperar evidencia textual de manera sistemática, mientras que los LLMs pueden ayudar a interpretar y sintetizar esa evidencia.
