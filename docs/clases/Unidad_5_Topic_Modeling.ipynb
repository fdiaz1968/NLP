## Desarrollo

### R — LDA con `topicmodels`

```r
library(quanteda)
library(topicmodels)

textos <- c(
  doc1 = "El Banco Central subió la tasa de interés ante el aumento de la inflación.",
  doc2 = "Los inversionistas reaccionaron con cautela frente a la volatilidad del mercado.",
  doc3 = "La economía mostró señales de recuperación durante el último trimestre."
)

dtm <- corpus(textos) |>
  tokens(remove_punct = TRUE) |>
  tokens_remove(stopwords("spanish")) |>
  dfm() |>
  convert(to = "topicmodels")

lda_model <- LDA(dtm, k = 3, control = list(seed = 123))
terms(lda_model, 5)
```

### Python — BERTopic

````python
from bertopic import BERTopic

documentos = [
    "El Banco Central subió la tasa de interés ante el aumento de la inflación.",
    "Los inversionistas reaccionaron con cautela frente a la volatilidad del mercado.",
    "La economía mostró señales de recuperación durante el último trimestre.",
]

modelo_bertopic = BERTopic(language="spanish")
topicos, probabilidades = modelo_bertopic.fit_transform(documentos)
modelo_bertopic.get_topic_info()
````

### Python — Etiquetado de tópicos con un LLM

````python
from transformers import pipeline

etiquetador = pipeline("text-generation", model="datificate/gpt2-small-spanish")

palabras_topico = ["inflación", "tasa", "banco", "central"]
prompt = "Genera una etiqueta corta para un tópico definido por: " + ", ".join(palabras_topico)

etiqueta = etiquetador(prompt, max_new_tokens=15)[0]["generated_text"]
print(etiqueta)
````
