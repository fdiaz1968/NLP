## Desarrollo

### Construcción de la DFM

```r
library(quanteda)

textos <- c(
  doc1 = "El Banco Central subió la tasa de interés ante el aumento de la inflación.",
  doc2 = "Los inversionistas reaccionaron con cautela frente a la volatilidad del mercado.",
  doc3 = "La economía mostró señales de recuperación durante el último trimestre."
)

corpus_nlp <- corpus(
  textos,
  docvars = data.frame(fuente = c("prensa", "prensa", "informe"))
)

toks <- tokens(corpus_nlp, remove_punct = TRUE) |>
  tokens_remove(stopwords("spanish"))

dfm_nlp <- dfm(toks)
dfm_nlp
```

### Ponderación, trimming y agrupación

```r
# Ponderación TF-IDF
dfm_tfidf_nlp <- dfm_tfidf(dfm_nlp)

# Trimming: conservar términos presentes en al menos 2 documentos
dfm_trim_nlp <- dfm_trim(dfm_nlp, min_docfreq = 2)

# Agrupar documentos por variable de metadatos
dfm_grouped <- dfm_group(dfm_nlp, groups = fuente)

# Exportar a data.frame para uso en modelos externos
df_tfidf <- convert(dfm_tfidf_nlp, to = "data.frame")
head(df_tfidf)
```
