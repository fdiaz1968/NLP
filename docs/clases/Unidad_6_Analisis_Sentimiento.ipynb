## Desarrollo

### Clasificación con FinBERT

```python
from transformers import pipeline

clasificador = pipeline("sentiment-analysis", model="ProsusAI/finbert")

noticias = [
    "Company profits soared after the earnings report.",
    "The firm faces a significant liquidity crisis.",
    "Markets remained flat amid mixed economic signals.",
]

resultados = clasificador(noticias)
for noticia, resultado in zip(noticias, resultados):
    print(f"{resultado['label']:>10} ({resultado['score']:.2f})  {noticia}")
```

### Evaluación con métricas de clasificación

```python
from sklearn.metrics import classification_report, confusion_matrix

etiquetas_reales = ["positive", "negative", "neutral"]
predicciones = [r["label"].lower() for r in resultados]

print(classification_report(etiquetas_reales, predicciones))
print(confusion_matrix(etiquetas_reales, predicciones, labels=["positive", "negative", "neutral"]))
```
